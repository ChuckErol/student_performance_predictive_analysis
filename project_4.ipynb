```python
import pandas as pd

#for the ML model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
```

```python
# data processing section
```

```python
df = pd.read_csv('Resourse/Student_performance_data _.csv')
df_model = df.copy()
df
```

```python
df = df.drop(columns=['StudentID'])
```

```python
df['Gender'] = df['Gender'].replace({0: 'male', 1: 'female'})
df
```

```python
df['Ethnicity'] = df['Ethnicity'].replace({0: 'caucasian', 1: 'african-american', 2: 'asian', 3: 'other'})
df
```

```python
df['ParentalEducation'] = df['ParentalEducation'].replace({0: 'none', 1: 'high-school', 2: 'some-college', 3: 'bachelors', 4: 'higher'})
df
```

```python
df['ParentalSupport'] = df['ParentalSupport'].replace({0: 'none', 1: 'low', 2: 'moderate', 3: 'high', 4: 'very high'})
df
```

```python
df['Tutoring'] = df['Tutoring'].replace({0: 'false', 1: 'true'})
df['Extracurricular'] = df['Extracurricular'].replace({0: 'false', 1: 'true'})
df['Sports'] = df['Sports'].replace({0: 'false', 1: 'true'})
df['Music'] = df['Music'].replace({0: 'false', 1: 'true'})
df['Volunteering'] = df['Volunteering'].replace({0: 'false', 1: 'true'})
df
```

```python
df['GradeClass'] = df['GradeClass'].replace({0.0: 'A', 1.0: 'B', 2.0: 'C', 3.0: 'D', 4.0: 'F'})
df
```

```python
df.to_csv('Resourse/updated_data.csv', index=False)
```

```python
# Model Training
```

```python
#Drop some columns
df_model = df_model.drop(columns=['StudentID', 'GradeClass'])
df_model.head()
```

```python
model_data_scaled = StandardScaler().fit_transform(df_model[['Age', 'StudyTimeWeekly', 'Absences']])

model_data_scaled[0:5]
```

```python
# Create a DataFrame called with the scaled data
# The column names should match those referenced in the StandardScaler step
df_model_scaled = pd.DataFrame(
    model_data_scaled,
    columns=['Age', 'StudyTimeWeekly', 'Absences']
)

# Review the DataFrame
df_model_scaled.head()
```

```python
#drop the newly scaled columns and add the new scaled columns back in their place
df_scaled = df_model.drop(columns=['Age', 'StudyTimeWeekly', 'Absences'])

# df_scaled = pd.concat([df_scaled, df_model_scaled])

df_scaled['Age'] = df_model_scaled['Age']
df_scaled['StudyTimeWeekly'] = df_model_scaled['StudyTimeWeekly']
df_scaled['Absences'] = df_model_scaled['Absences']
# df_scaled['GPA'] = df_model_scaled['GPA']

df_scaled.head()
```

```python
#Drop the target column
y = df_model['GPA']

X = df_model.drop(columns=['GPA'])
```

```python
#Split the data for training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
#Instantiate some models
models = []

#Logistic Regression Model
linear_regression_model = LinearRegression() #Classification, random forest regressor, logistical regression is a classification model
# models.append(logistic_regression_model)

# K Means Model
# k_means = KMeans(n_clusters=2)
# models.append(k_means)

# models
```

```python
linear_regression_model.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {linear_regression_model.score(X_train, y_train)}")
print(f"Testing Data Score: {linear_regression_model.score(X_test, y_test)}")
```

```python
# Predict outcomes for test data set
predictions = linear_regression_model.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```
